```python
import numpy as np
import pandas as pd

load_data = np.load("new_processed_dataset.npz")

x = np.concatenate([
    load_data["X_train"],
    load_data["X_val"],
    load_data["X_test"]
])
y = np.concatenate([
    load_data["y_train"],
    load_data["y_val"],
    load_data["y_test"]
])

x_flat= x.reshape(x.shape[0],-1)
df = pd.DataFrame(x_flat)
df["label"]=y

df.to_csv("processed_data.csv",index= False)
print("CSV saved")
print("shape:", df.shape)
```
