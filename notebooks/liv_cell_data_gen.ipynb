```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = np.zeros((512, 512, 3), np.uint8)
cv2.circle(
    img,(32,32), 20, (180, 160, 100), -1
)
cv2.circle(
    img,(32,32), 8, (60,20,80), -1
)
plt.imshow(img)
plt.show()
```

```python
radius = np.random.randint(18,24)

img = np.zeros((512,512,3), dtype=np.uint8)

cv2.circle(
    img,
    (32,32),
    radius,
    (180,160,100),
    -1
)
cv2.circle (
    img,
    (32,32),
    radius//2,
    (60,20,80),
    -1
)
plt.imshow(img)
plt.show()
```

```python
noise = np.random.normal(
    0,
    10,
    img.shape
)

img = np.clip(
    img.astype(np.int16) + noise,
    0,
    255
).astype(np.uint8)
plt.imshow(img)
plt.show()
```

```python
def generate_healthy():
    
    img = np.zeros((64,64,3), dtype=np.uint8)

    cyto_radius = np.random.randint(20,25)

    cv2.circle(
        img,
        (32,32),
        cyto_radius,
        (180,160,100),
        -1
    )

    nucleus_radius = np.random.randint(7,10)

    cv2.circle(
        img,
        (32,32),
        nucleus_radius,
        (60,20,80),
        -1
    )

    noise = np.random.normal(
        0,
        8,
        img.shape
    )

    img = np.clip(
        img.astype(np.int16)+noise,
        0,
        255
    ).astype(np.uint8)

    return img
```

```python
fig, axes = plt.subplots(3,4, figsize=(4,4))

for ax in axes.ravel():

    img = generate_healthy()

    ax.imshow(img)
    ax.axis("off")

plt.show()
```

```python
def generate_damage():
    
    img1 = np.zeros((64,64,3), dtype=np.uint8)
    ax=np.random.randint(14,20)
    ay = np.random.randint(16,22)
    # cyto_radius = np.random.randint(20,25)

    cv2.ellipse(
        img1,
        (32,32),
        (ax,ay),
        # cyto_radius,
        0,0,360,
        (180,160,100),
        -1
    )

    # nucleus_radius = np.random.randint(7,10)
    bx = np.random.randint(5,10)
    by = np.random.randint(7,12)
    cv2.ellipse(
        img1,
        (32,32),
        (bx,by),
        # nucleus_radius,
        0,0,360,
        (60,20,80),
        -1
    )

    noise = np.random.normal(
        4,
        12,
        img1.shape
    )

    img1 = np.clip(
        img1.astype(np.int16)+noise,
        0,
        255
    ).astype(np.uint8)
    

    return img1

fig, axes = plt.subplots(3,4, figsize=(4,4))

for ax in axes.ravel():

    img1 = generate_damage()

    ax.imshow(img1)
    ax.axis("off")

plt.show()
```
